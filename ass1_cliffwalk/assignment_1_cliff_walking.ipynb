```python
# Helper functions for implementing Q_Learning and SARSA


def init_Q(grid_size): # Initialize a Q table
    Q = {}
    for x, y in np.ndindex((grid_size, grid_size)):
        for action in ['UP', 'DOWN', 'LEFT', 'RIGHT']:
            Q[((x,y),action)] = random.random()
    return Q

def argmax_Q(Q, x, y): # Return the maximum action corresponding to a state
    D = {}
    for action in ['UP', 'DOWN', 'LEFT', 'RIGHT']:
        D[action] = Q[((x,y), action)]
    return max(D, key=D.get)

def e_greedy(epsilon, x, y, Q):  # Return the e-greedy action from a Q table
    if random.random()< epsilon:
        return random.choice(['UP', 'DOWN', 'LEFT', 'RIGHT'])
    else:
        return argmax_Q(Q, x, y)

def plot_policy(Q, grid_size): # Plot the policy 
    policy = np.chararray((grid_size, grid_size), unicode=True)
    for x, y in np.ndindex((grid_size, grid_size)):
        policy[x,y] =  argmax_Q(Q, x, y)
    policy[0,1:-1] = 'H'

    policy[0, grid_size - 1] = 'C'
    return policy
        
```
